```python
import re
def extract_ingredient_names(ingredient):
    # Remove any leading numbers, fractions, and units
    cleaned_ingredient = re.sub(r'^\s*\d+(?:/\d+)?\s*(?:cups?|grams?|g|kg|tablespoons?|tbsp|teaspoons?|tsp|ml|l|ounces?|oz|pounds?|lb|pinch|dash|cloves?|pieces?|inch|sprig|cup|tablespoon|teaspoon)?\s*', '', ingredient, flags=re.IGNORECASE)
    # Remove any descriptions in parentheses or after the main name
    #cleaned_ingredient = re.sub(r'\s*\(.*?\)\s*', '', cleaned_ingredient).strip()
    # Remove any trailing descriptors like "finely chopped", "washed", etc.
    cleaned_ingredient = re.sub(r',.*$', '', cleaned_ingredient).strip()
    return cleaned_ingredient
```

```python
cleaned_ingredients = [extract_ingredient_names(ingredient) for ingredient in df['final_ingredients']]
```
