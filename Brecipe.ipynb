```python
df = pd.read_csv('recipe.csv')
```

```python
rows = df.values.tolist()
wrapped_rows = [f"[{','.join(map(str, row))}]" for row in rows]
wrapped_rows
```

```python
df_wrapped = pd.DataFrame(wrapped_rows, columns=['wrapped_row'])
df_wrapped.to_csv('Brecipe.txt', index=False, header=False)
```
